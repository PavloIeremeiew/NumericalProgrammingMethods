```python
import math
import numpy as np
from scipy.misc import derivative
import numdifftools as nd
```

```python
def f(x):#функція
  return 3*pow(x, 4) + 4*pow(x,3) - 12*pow(x, 2) - 5
eps=0.0001
```

```python
def findSegments():
  searchRange = np.arange(-5, 5, 1)#область для пошуку можливих x
  a = None
  previousFunc = None
  currentFunc = None
  segments = []# масив значень х
  for x in searchRange:# цикл області область значень
    x = round(x, 4)# округлення похибки
    currentFunc= f(x)
    if previousFunc != None and previousFunc * currentFunc< 0:# перевірка на зміну знаку функції
      segments.append((a, x))#створення пари
    a = x
    previousFunc = currentFunc
  return segments
```

```python
print(findSegments())
```

```python
def halfDivision(a,b,e):# метод половинного ділення
  n=(a+b)/2
  if(abs(a-b)>e):
    if(f(a)*f(n)<0):
      return halfDivision(a,n,e)
    else:
      return halfDivision(n,b,e)

  return  round(n,5)
```

```python
print("х1:",halfDivision(-3,-2,eps))# знайдемо х1 метод половинного ділення
print("х2:",halfDivision(1,2,eps))# знайдемо х2 метод половинного ділення
```

```python
def Chord(a,b,e,x0_is_b=False):# метод хорд
  x=a-f(a)*(b-a)/(f(b)-f(a))
  x0= b if x0_is_b  else a
  if(abs(x-x0)<e):
    return round(x,5)
  if(f(a)*f(x)<0):
    return Chord(a,x,e,True)
  return Chord(x,b,e,False);
```

```python
print("х1:",Chord(-3,-2,eps))# знайдемо х1 метод хорд
print("х2:",Chord(1,2,eps))# знайдемо х2 метод хорд
```
